## Problem 2

```python
def problem2(p, d=100):
    start = "1/1/2000"
    end = "12/31/2000"
    dates = pd.date_range(start, end, freq='D')
    
    np.random.binomial(1, p, size=len(dates))
    stock_vals = np.random.binomial(1, p, size=len(dates))
    stock_vals[stock_vals == 0] = -1
    stock_vals[0] = d    
    stock_series = pd.Series(stock_vals, index=dates, name='Stock')
    stock_series = stock_series.cumsum()
    stock_series[stock_series < 0] = 0
    stock_series.plot(title="Stock Value")
    plt.ylabel("Stock Value")
    plt.xlabel("Date")
```

```python
problem2(0.5)
```

```python
problem2(0.3)
```

```python
problem2(0.8)
```
