# Pandas 2: Plotting

```python
from pydataset import data
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats, special
```

### Road Accident Deaths in the US

```python
road = data("road")
#data("road", show_doc=True)
```

```python
road.head()
```

```python
pd.plotting.scatter_matrix(road[["deaths", "popden", "rural", "temp"]])
plt.annotate(r"Source: Imperial College, London M.Sc. exercise", (-3,0), (0, -60), xycoords='axes fraction', textcoords='offset points', va='top')
```

This gives us a comprehensive understanding of possible correlation between different variables.

```python
road.plot(kind="scatter", x="rural", y="deaths", title="Rural vs Deaths")
slope, intercept, r_value, p_value, std_err = stats.linregress(road.rural, road.deaths)
plt.plot(road.rural, intercept + slope * road.rural, \
          'r-', label=r"Fitted Line ($R^2 = {0:.2f}$)".format(r_value))
plt.legend()

plt.annotate(r"Source: Imperial College, London M.Sc. exercise", \
             (0,0), (0, -40), xycoords='axes fraction', textcoords='offset points', va='top')
```

Given the comprehensive summary, we decide to look closer in the correlation between rural areas and deaths. The linear regression seems to fit well at first sight, but the $R^2$ is pretty low.

```python
road[["deaths"]].plot(kind="barh", title="Road Accident Deaths in US")
plt.annotate(r"Source: Imperial College, London M.Sc. exercise", (0,0), (0, -30), xycoords='axes fraction', textcoords='offset points', va='top')
```

```python
road[["rural"]].plot(kind="barh", title="Length of Rural Roads")
plt.annotate(r"Source: Imperial College, London M.Sc. exercise", (0,0), (0, -30), xycoords='axes fraction', textcoords='offset points', va='top')
```

Based on the comprehensive visualization among different variables, there seemed to be a correlation between road accident deaths and length of rual roads. However, further investigation seems to suggest otherwise.  

### Risk Factors Associated with Low Infant Birth Weight

```python
Arbuthnot=data("Arbuthnot")
#data("Arbuthnot", show_doc=True)
```

```python
Arbuthnot.head()
```

```python
Arbuthnot.plot(x='Year', y='Ratio', title="Ratio of Males/Females in London")
annotate_string = r"Source: Arbuthnot, John (1710)."
annotate_string += "\n"
annotate_string += r"'An argument for Devine Providence, taken from theconstant Regularity observ'd in the Births of both Sexes'"
annotate_string += "\n"
annotate_string += r"_Philosophicaltransactions_, 27, 186-190. Published in 1711."
plt.annotate(annotate_string, (0,0), (0, -40), xycoords='axes fraction', textcoords='offset points', va='top')
```

```python
Arbuthnot.plot(x='Year', y=['Mortality', 'Plague'], title="Total Mortality (with Plague Mortality) in London")
plt.ylabel("Count")
annotate_string = r"Source: Arbuthnot, John (1710)."
annotate_string += "\n"
annotate_string += r"'An argument for Devine Providence, taken from theconstant Regularity observ'd in the Births of both Sexes'"
annotate_string += "\n"
annotate_string += r"_Philosophicaltransactions_, 27, 186-190. Published in 1711."
plt.annotate(annotate_string, (0,0), (0, -40), xycoords='axes fraction', textcoords='offset points', va='top')
```

This graph seems to explain high correlation between plague and mortality in London.

```python
Arbuthnot.plot(kind="scatter", x='Ratio', y='Mortality', title="Ratio of Males/Females vs. Mortality in London")
annotate_string = r"Source: Arbuthnot, John (1710)."
annotate_string += "\n"
annotate_string += r"'An argument for Devine Providence, taken from theconstant Regularity observ'd in the Births of both Sexes'"
annotate_string += "\n"
annotate_string += r"_Philosophicaltransactions_, 27, 186-190. Published in 1711."
plt.annotate(annotate_string, (0,0), (0, -40), xycoords='axes fraction', textcoords='offset points', va='top')
```

This graph shows that there is no correlation between mortality and males/females ratio in England. In fact, total mortality seems to be somewhat consistent in the range of 15000 with noise.
