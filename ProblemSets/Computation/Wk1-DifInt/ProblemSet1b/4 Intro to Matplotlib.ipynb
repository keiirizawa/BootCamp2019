# Intro to Matplotlib

```python
import numpy as np 
from matplotlib import pyplot as plt
```

## Problem 1

```python
def prob1_1(n):
    sample = np.random.normal(size=(n, n))
    mean = np.mean(sample, axis=1)
    var = np.var(mean)    
    return var
```

```python
def prob1_2():    
    n_grid = np.arange(100, 1000, 100)
    calc = np.vectorize(prob1_1)
    results = calc(n_grid)
    plt.plot(n_grid,results)
    plt.show()
```

```python
prob1_2()
```

## Problem 2

```python
def prob2():
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    plt.plot(x, np.sin(x))
    plt.plot(x, np.cos(x))
    plt.plot(x, np.arctan(x))
    plt.show()
```

```python
prob2()
```

## Problem 3

```python
def f(x):
    return 1 / (x - 1)
```

```python
def prob3():
    domain_l = np.linspace(-2,1-0.05, 50)
    domain_r = np.linspace(1+0.05, 6, 50)
    
    plt.plot(domain_l, f(domain_l), "m--", lw=4)
    plt.plot(domain_r, f(domain_r), "m--", lw=4)
    plt.xlim(-2,6)
    plt.ylim(-6,6)
    plt.show()
```

```python
prob3()
```

## Problem 4

```python
def prob4():
    
    x = np.linspace(0, 2*np.pi, 100)
    
    fig, axes = plt.subplots(2,2)
    axes[0,0].plot(x, np.sin(x), "g-")
    axes[0,0].set_title(r'$sin(x)$')
    axes[0,1].plot(x, np.sin(2*x), "r-")
    axes[0,1].set_title(r'$sin(2x)$')
    axes[1,0].plot(x, 2*np.sin(x), "b--")
    axes[1,0].set_title(r'$2sin(x)$')
    axes[1,1].plot(x, 2*np.sin(2*x), "m:")
    axes[1,1].set_title(r'$2sin(2x)$')
    
    plt.suptitle('Functions')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.tight_layout()
    plt.show()
    
```

```python
prob4()
```

## Problem 5

```python
def prob5():
    fars = np.load('FARS.npy')
    time, long, lat = fars.T
    
    fig = plt.figure(figsize=(12,5))
    
    ax1 = plt.subplot(121)
    ax1.plot(long, lat, "k,")
    ax1.set_xlabel("Longitude")
    ax1.set_ylabel("Latitude")
    ax1.set_aspect("equal")

    ax2 = plt.subplot(122)
    ax2.hist(time, bins=24, range=[0,24])
    ax2.set_xlim(0,23)
    ax2.set_xlabel("Time (Hour)")
    ax2.set_ylabel("Frequency of Accidents")
    
    plt.suptitle("Summary Statistics of FARS")
    plt.tight_layout()
    plt.show()
```

```python
prob5()
```
