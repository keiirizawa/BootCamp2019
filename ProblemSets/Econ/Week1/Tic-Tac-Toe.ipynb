Assume that I am player O and giving an initial score to all possible grids (eliminates all impossible grids). Give a score of 1 if grid with 3 O in a row (win), give a score of -1 if grid with 3 X in a row (loss), give a score of 0 to rest.

```python
def is_legal_state(combo):
    
    curr_win_pos = {None}
    strike_count = 0
    win = False 

    counter=collections.Counter(combo)

    if abs(counter[1] - counter[2]) >= 2:
        return False, None

    elements = [0,1,2]
    if combo[0] == combo[4] == combo[8] != 0:
        
        curr_win_pos = {0,4,8}
        
        if combo[0] == 1:
            win = True
       
        # freq diff btw O and X
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[2] == combo[4] == combo[6] != 0:
        
        if combo[2] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[2] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({2,4,6})) == 1:
                        return False, None
                    
            else:
                if combo[2] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({2,4,6})) == 1:
                        return False, None
        
        else:
            curr_win_pos = {2,4,6}
        
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[0] == combo[1] == combo[2] != 0:
        
        if combo[0] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[0] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({0,1,2})) == 1:
                        return False, None
                    
            else:
                if combo[0] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({0,1,2})) == 1:
                        return False, None
        else:
            curr_win_pos = {0,1,2}
        
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[3] == combo[4] == combo[5] != 0:
        
        if combo[3] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[3] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({3,4,5})) == 1:
                        return False, None
                    
            else:
                if combo[3] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({3,4,5})) == 1:
                        return False, None
        else:
            curr_win_pos = {3,4,5}
            
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[6] == combo[7] == combo[8] != 0:
        
        if combo[6] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[6] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({6,7,8})) == 1:
                        return False, None
                    
            else:
                if combo[6] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({6,7,8})) == 1:
                        return False, None
        
        else:
            curr_win_pos = {6,7,8}
        
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[0] == combo[3] == combo[6] != 0:
        
        if combo[0] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[0] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({0,3,6})) == 1:
                        return False, None
                    
            else:
                if combo[0] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({0,3,6})) == 1:
                        return False, None
        
        else:
            curr_win_pos = {0,3,6}
            
            
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[1] == combo[4] == combo[7] != 0:
        
        if combo[1] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[1] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({1,4,7})) == 1:
                        return False, None
                    
            else:
                if combo[1] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({1,4,7})) == 1:
                        return False, None
        
        else:
            curr_win_pos = {1,4,7}
        
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    elements = [0,1,2]
    if combo[2] == combo[5] == combo[8] != 0:
        
        if combo[2] == 1:
            win = True
        
        if strike_count == 1:
            if win == True:
                if combo[2] == 2:    
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({2,5,8})) == 1:
                        return False, None
                    
            else:
                if combo[2] == 1: 
                    return False, None
                else:
                    if not len(curr_win_pos.intersection({2,5,8})) == 1:
                        return False, None
        
        else:
            curr_win_pos = {2,5,8}
        
        if win:
            if counter[1] - counter[2] != 1:
                return False, None
        else:
            if counter[1] - counter[2] != 0:
                return False, None

        strike_count += 1

    if strike_count == 1:
        if win:
            return True, 1
        else:
            return True, -1
    else:
        return True, 0
```

```python
def initial_setup(): 
    states_score = dict()
    for combo in itertools.product([0, 1, 2], repeat=9):
        
        is_legal, score = is_legal_state(combo)
        
        if is_legal:
            states_score[combo] = score
    
    return states_score
```

```python
def checking_if_terminal(combo):
    if combo[0] == combo[3] == combo[6] != 0:
        if combo[0] == 1:
            return True, 1
        else:
            return True, -1
    
    if combo[0] == combo[1] == combo[2] != 0:
        if combo[0] == 1:
            return True, 1
        else:
            return True, -1
        
    if (combo[0] == combo[3] == combo[6] != 0):
        if combo[0] == 1:
            return True, 1
        else:
            return True, -1
        
    if combo[2] == combo[4] == combo[6] != 0:
        if combo[2] == 1:
            return True, 1
        else:
            return True, -1
        
    if combo[3] == combo[4] == combo[5] != 0:
        if combo[3] == 1:
            return True, 1
        else:
            return True, -1
        
    if combo[1] == combo[4] == combo[7] != 0:
        if combo[4] == 1:
            return True, 1
        else:
            return True, -1
    
    if combo[6] == combo[7] == combo[8] != 0:
        if combo[6] == 1:
            return True, 1
        else:
            return True, -1
        
    if combo[0] == combo[3] == combo[6] != 0:
        if combo[6] == 1:
            return True, 1
        else: 
            return True, -1
    
    if combo[1] == combo[4] == combo[7] != 0:
        if combo[1] == 1:
            return True, 1
        else: 
            return True, -1
    
    if combo[2] == combo[5] == combo[8] != 0:
        if combo[1] == 1:
            return True, 1
        else: 
            return True, -1  
    
    else:
        return False, 0
```

Assume I make first move:

```python
def iterate(states_score):
    
    for combo, v in states_score.items():
        counter=collections.Counter(combo)

        is_terminal, score = checking_if_terminal(combo)
        if is_terminal:
            states_score[combo] = score
        
        else:
            
            if counter[1] > counter[2]: # this is opponent's turn
                states_score[combo] = opp_value_fun(states_score, combo)
            if counter[1] == counter[2]: # this is my turn
                states_score[combo] = my_value_fun(states_score, combo)
                
    return states_score
```

```python
def my_value_fun(states_score, curr_state):
    
    best_score = states_score[curr_state]
    for pos in range(len(curr_state)):
        if curr_state[pos] == 0:  
            copy_state = list(curr_state)
            copy_state[pos] = 1
            is_legal, score = is_legal_state(copy_state)
            if is_legal:
                score = states_score[tuple(copy_state)]
                if score > best_score:
                    best_score = score
        
    return best_score
```

```python
def opp_value_fun(states_score, curr_state):
    
    worst_score = states_score[curr_state]
    for pos in range(len(curr_state)):
        if curr_state[pos] == 0:  
            copy_state = list(curr_state)
            copy_state[pos] = 2  
            is_legal, score = is_legal_state(copy_state)
            if is_legal:
                score = states_score[tuple(copy_state)]
                if score < worst_score:
                    worst_score = score
        
    return worst_score
```

```python
def calc_diff(states_score, updated_states_score):
    diff_sum = 0
    for k, v in states_score.items():
        diff = abs(updated_states_score[k] - v)
        diff_sum += diff
    return diff_sum
```

```python
def value_func_iteration(threshold):
    states_score = initial_setup()
    
    diff = 100
    iter_ = 1
    while diff > threshold: 
        updated_states_score = iterate(states_score.copy())
        diff = calc_diff(states_score, updated_states_score)
        states_score = updated_states_score.copy()
        iter_ += 1
        
    return states_score
```
