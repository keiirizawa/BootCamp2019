# Section 2: Combinatorial Choice Problem

## Problem (a)

The profit function we consider is from Antras, Fort, and Tintelnot (2014):
$$\pi(J, \varphi, f_{ij}^n) = \varphi^{\sigma-1} \left( \sum_{j=1}^{j \in J} T_j (\tau_{ij}w_{j})^{-\theta} \right)^{\sigma-1/\theta} B - \sum_{j \in J} f_{ij}^n$$
where $\phi$: productivity, $f_{ij}^n$: fixed cost 

## Problem (b)

Parameters of the model come from calibration values and methods from Antras, Fort, and Tintelnot (2014):

- $\xi_{j} = T_j (\tau_{ij}w_{j})^{-\theta}$ 

- $\log \xi_{j}=\beta_{r} \log \mathrm{R} \& \mathrm{D}_{j}+\beta_{k} \log \operatorname{capital}_{j}+\beta_{F} \log \text { number of firms }_{j} +\beta_{C} \text { control of corruption }_{j}  -\theta \log w_{j}-\theta\left(\log \beta_{c}+\beta_{d} \log \operatorname{distance}_{i j}+\operatorname{language}_{i j} \log \beta_{l}\right. )+\iota_{j}$

- $\beta_0 = -7.250, \beta_r = 0.352, \beta_k = -0.184, \beta_F = 0.108, \theta = -0.537, \beta_d = -0.341, \beta_l = 0.105, \beta_C = 0.105 $ 

- $B$ = 0.122

- $f_{ij}^n$ drawn from lognormal distribution with dispersion parameter $\beta_{\text{disp}}^f$ and scale parameter $\log \beta_{c}+\beta_{d}^{f} \log \text { distance }_{i j} + \log \beta_{t}^{f} \text { language }_{i j}+\beta_{C}^{f} \text { control of corruption }_j$
- $\beta_c^f = 0.022, \beta_d^f = 0.193, \beta_l^f = 0.872, \beta_C^f = -0.393, \beta_{\text{disp}}^f = 0.934$

- $\frac{\sigma-1}{\theta} < 1$ (Substitution case)

#### Ekchert et al. (2017) Algorithm

##### Marginal benefit of adding country j given $\phi$ and $J$ \ $j$
$$\varphi^{\sigma-1} B\left(\Theta_{i}(\mathcal{J})^{(\sigma-1) / \theta}-\Theta_{i}(\mathcal{J} \backslash j)^{(\sigma-1) / \theta}\right)-f_{i j}^{n}$$

where $\Theta_{i} = \sum_{k \in J_i(\phi)} T_k (\tau_{ik}w_k)^{-\theta}$.

Therefore, $\left(\Theta_{i}(\mathcal{J})^{(\sigma-1) / \theta}-\Theta_{i}(\mathcal{J} \backslash j)^{(\sigma-1) / \theta}\right) = \hat{\xi_j}$.

We will set $\varphi = 1$, $\sigma=1.5$, $\theta=1.789$ (Benchmark Case).

```python
def get_xi(index):
    
    β_0 = -7.250
    β_r = 0.352
    β_k = -0.184
    β_F = 0.108
    θ = -0.537
    β_d = -0.341
    β_C = 0.105
    
    c_df = df.loc[index]
    
    xi = β_0 + β_r*c_df['log_rd'] + β_k*c_df['log_KL'] + β_F*c_df['log_firms'] \
        + β_C*c_df['control_of_corruption'] - θ*c_df['log_wage'] \
        - θ*( β_d*c_df['log_dist'])
    
    return xi
```

```python
def get_fij(index):
    
    β_c = 0.022
    β_d = 0.193
    β_l = 0.872
    β_C = -0.393
    β_disp = 0.934
    
    c_df = df.loc[index]
    
    dispersion = β_disp
    scale = np.log(β_c) + β_d*np.log(c_df['log_dist']) + np.log(β_l)*c_df['ls'] \
                + β_C*c_df['control_of_corruption']

    return np.random.lognormal(dispersion, scale, 1)
```

```python
def marginal_benefit(index, add=True):
    φ = 1
    B = 0.122
    ξ = get_xi(index)
    fij = get_fij(index)
    
   
    mb = φ * B * ξ - fij
       
    if add:
        return mb
    else:
        return -mb        
```

```python
def profit(J):
    φ = 1
    B = 0.122
    σ = 1.5
    θ = -0.537
    
    theta_sum = 0
    f_sum = 0
    for i in range(len(J)):
        if J[i] == 1:
            theta_sum += get_xi(i)
            f_sum += get_fij(i)
            
    pi = φ * B * ( theta_sum ** ((σ-1)/θ) ) - f_sum
    
    return pi    
```

```python
def ekchert_alg(N):
    
    ## Initialize J_vectors:
    J_min = np.zeros(df.country.shape[0])
    J_max = np.ones(df.country.shape[0])
    
    for i in range(N):
        index = random.randint(0, 239)
        if J_min[index] == 0:
            mb = marginal_benefit(index, add=True)
            if mb > 0:
                J_min[index] == 1
        else:
            mb = marginal_benefit(index, add=False)
            if mb > 0:
                J_min[index] == 0
        
        if J_max[index] == 0:
            mb = marginal_benefit(index, add=True)
            if mb > 0:
                J_max[index] == 1
        else:
            mb = marginal_benefit(index, add=False)
            if mb > 0:
                J_max[index] == 0
    
    ### Choose J with bigger profit:
    pi_min = profit(J_min)
    pi_max = profit(J_max)
    
    if pi_min > pi_max:
        return J_min
    else:
        return J_max
```
