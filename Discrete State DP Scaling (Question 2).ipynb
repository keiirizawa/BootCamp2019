# Discrete State DP Scaling (Question 2)

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Time it takes to run value function iteration for 30 MPI nodes and varying CPU threads

```python
CPU_count = np.linspace(1, 15, 15)
CPU_times = np.array([149.227374, 77.692148, 55.368205, 43.308273, 36.404679, 30.584016, 27.072396, 23.374039,
                     20.656425, 18.756201, 17.871441, 16.050265, 15.209664, 14.421857, 17.381238])
CPU_times_norm = CPU_times / CPU_times[0]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(CPU_count, CPU_times_norm)
ax.set_xlabel('CPU count')
ax.set_xticks(CPU_count)
ax.set_ylabel('Runtime')
ax.set_title('Scaling Plots Normalized to 1 CPU Thread (8 MPI nodes used)')
ax.grid()
plt.tight_layout()
plt.show();
```

### Time it takes to run value function iteration for 28 CPU threads and varying MPI nodes 

```python
node_count = np.linspace(1, 9, 9)
node_times = np.array([61.223205, 32.164540, 23.124772, 17.542765, 14.907700, 14.326791,
                      15.261105, 14.234768, 14.212368])
node_times_norm = node_times / node_times[0]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(node_count, node_times_norm)
ax.set_xlabel('Node count')
ax.set_xticks(node_count)
ax.set_ylabel('Runtime')
ax.set_title('Scaling Plots Normalized to 1 MPI node (28 CPU Threads used)')
ax.grid()
plt.tight_layout()
plt.show();
```
